Preprocesar los datos


```
# cambiar los valores en la columna de Sexo
train['Sex'].replace(['female','male'],[0,1],inplace=True)
test['Sex'].replace(['female','male'],[0,1],inplace=True)
```

```
# cambiar los valores de embarque
train['Embarked'].replace(['Q','S','C'],[0,1,2],inplace=True)
test['Embarked'].replace(['Q','S','C'],[0,1,2],inplace=True)
```

```
# completar valores en edad
promedio = 30
train['Age'] = train['Age'].replace(np.nan, promedio)
test['Age'] = train['Age'].replace(np.nan, promedio)
```

```
# crear grupos de edades
bins = [0, 8, 15, 18, 25, 40, 60, 80]
nombres = ['1', '2', '3', '4', '5', '6', '7']
train['Age'] = pd.cut(train['Age'], bins, labels=nombres)
test['Age'] = pd.cut(test['Age'], bins, labels=nombres)
```

```
# eliminar la columna Cabina
train.drop(['Cabin'], axis=1, inplace=True)
test.drop(['Cabin'], axis=1, inplace=True)
```

```
# eliminar columnas que se consideran sin valor agregado
train = train.drop(['PassengerId', 'Name', 'Ticket'], axis=1)
test = test.drop(['Name', 'Ticket'], axis=1)
```

```
# eliminar datos faltantes
train.dropna(axis=0, how='any', inplace=True)
test.dropna(axis=0, how='any', inplace=True)
```

Algoritmos de Machine Learning

1- separar la informacion de sobrevivientes
2- separar los datos de entrenamiento y test
# algoritmos a usar
3- Regresion logistica
4- soporte de maquinas  vectores
4 Vecinos proximos

```
from sklearn.model_selection import train_test_split # split de preubas
from sklearn.linear_model import LogisticRegression # regresion logistica
from sklearn.svm import SVC # soporte de maquinas vectores
from sklearn.neighbors import KNeighborsClassifier # vecinos proximos
```

```
# separamos la informacion de sobrevivientes
X = np.array(train.drop(['Survived'],1))
y = np.array(train['Survived'])
```

```
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```
# regresion logistica
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
Y_predict = logreg.predict(X_test)
print('Precision de la Regresion Logistica')
logreg.score(X_train,y_train)
```

```
# soporte de maquinas de vectores
svc = SVC()
svc.fit(X_train,y_train)
Y_pred = svc.predict(X_test)
print('Precision de soporte de maquina de vectores')
svc.score(X_train, y_train)
```

```
# vecinos proximos
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
Y_pre = knn.predict(X_test)
print('Precision Vecinos proximos')
knn.score(X_train,y_train)
```

Realizamos las predicciones con test

```
ids = test['PassengerId']
```

```
# regresion logistica
prediccion_logreg = logreg.predict(test.drop('PassengerId', axis=1))
salida_logreg = pd.DataFrame({'PassengerId': ids, 'Survived': prediccion_logreg})
print('Prediccion Regresion Logistica')
salida_logreg.head()
```

```
# soporte de maquinas de vectores
prediccion_svc = svc.predict(test.drop('PassengerId', axis=1))
salida_svc = pd.DataFrame({'PassengerId': ids, 'Survived': prediccion_svc})
print('Prediccion soporte de maquinas de vectores')
salida_svc.head()
```

```
# vecinos proximos
prediccion_knn = knn.predict(test.drop('PassengerId', axis=1))
salida_knn = pd.DataFrame({'PassengerId': ids, 'Survived': prediccion_knn})
print('Prediccion vecinos proximos')
salida_knn.head()
```
