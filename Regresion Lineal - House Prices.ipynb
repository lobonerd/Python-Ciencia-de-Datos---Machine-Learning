```
from sklearn.model_selection import train_test_split #separa en test y train
```

```
X = train.loc[:,["OverallQual", "GrLivArea", "GarageCars", "TotalBsmtSF", "YearBuilt", "YearRemodAdd" ]]
y = train.loc[:,["SalePrice"]]
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=33)
```

```
# importamos la regresion lineal
from sklearn.linear_model import LinearRegression
```

```
lr = LinearRegression()
```

```
# hacemos el fit, utilizar los datos de train
lr.fit(X_train,y_train)
```

```
print(lr.intercept_)
```

```
print(lr.coef_)
```

```
predicciones = lr.predict(X_test) # vamos a crear las predicciones
```

```
DTpredicciones = pd.DataFrame(predicciones) # convertimos las predicciones en dataframe de pandas
DTpredicciones.reset_index(drop= True, inplace= True) # reseteamos los indices
y_test.reset_index(drop= True, inplace= True) # reseteamos los indices
df_unido = y_test.join(DTpredicciones) # unimos los 2 dataframes
```

```
df_unido
```

```
# Vamos a sacar las metricas
from sklearn import metrics
```

```
# utilizamos el y_test y prediciones y comparar los valores reales de y_test contra predicciones
print('MAE:', metrics.mean_absolute_error(y_test, predicciones))
```

```
# utilizamos el y_test y prediciones y comparar los valores reales de y_test contra predicciones
print('MSE:', metrics.mean_squared_error(y_test, predicciones))
```

```
# utilizamos el y_test y prediciones y comparar los valores reales de y_test contra predicciones
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predicciones)))
```

```
# logaritmo RMSE
from sklearn.metrics import mean_squared_log_error
```

```
print('Log RMSE:', np.sqrt(metrics.mean_squared_log_error(y_test, predicciones)))
```
